Matplotlib: A versatile library for creating static, interactive, and animated visualizations. It provides low-level control over plots, enabling highly customized visualizations but requires more code to produce aesthetically pleasing graphs.

Seaborn: Built on top of Matplotlib, it simplifies the creation of complex visualizations and statistical plots with aesthetically pleasing defaults. It is specifically designed for statistical data visualization and works seamlessly with Pandas data structures.

1. Introduction to Matplotlib

Matplotlib is a robust and widely used Python library for data visualization. It is designed to generate
high-quality, publication-ready plots with ease. Whether you need a simple line chart or a complex 3D
plot, Matplotlib provides the flexibility to create static, animated, and interactive visualizations.

Why Use Matplotlib?

Versatility: Suitable for various types of plots like line charts, scatter plots, histograms, etc.
Customization: Offers detailed control over each plot element.
Integration: Works seamlessly with other Python libraries such as NumPy, pandas, and seaborn.

1. Introduction to Seaborn
Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface
for creating attractive and informative statistical graphics. It simplifies complex visualizations with
fewer lines of code and includes built-in themes and color palettes for enhanced aesthetic appeal.

Why Use Seaborn?

Ease of Use: Simplifies the creation of complex plots with simple functions.
Built-in Themes: Provides beautiful default styles and color palettes.
Integration with Pandas: Works seamlessly with DataFrame structures.
Statistical Plots: Designed specifically for data visualization that requires statistical analysis.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
help(plt)
```

```python
x = np.linspace(0, 20, 100)
y = np.sin(x)

plt.figure(figsize=(10, 5)) 
plt.plot(x, y, label='sin(x)', color='r') 

plt.title("sin curve")
plt.xlabel("X axis")
plt.ylabel("Y axix")

# Add a legend to identify the plotted line
plt.legend() 
# Automatically adjust spacing to prevent overlap of labels or titles
plt.tight_layout() 
# Add a grid to the plot for better readability
plt.grid(True) 

plt.show()
```

```python
x = np.random.rand(30)
y = np.random.rand(30)

plt.figure(figsize=(10, 8)) 
plt.scatter(x, y, label='sin(x)', color='r') 

plt.title("sin curve")
plt.xlabel("X axis")
plt.ylabel("Y axix")

# Add a legend to identify the plotted line
plt.legend() 
# Automatically adjust spacing to prevent overlap of labels or titles
plt.tight_layout() 
# Add a grid to the plot for better readability
plt.grid(True) 

plt.show()
```

```python
x = ["cats", "dogs", "birds"]
y = [10, 22, 5]

plt.figure(figsize=(5, 5)) 
plt.bar(x, y, label='sin(x)', color='r') 

plt.title("sin curve")
plt.xlabel("X axis")
plt.ylabel("Y axix")

# Add a legend to identify the plotted line
plt.legend() 
# Automatically adjust spacing to prevent overlap of labels or titles
plt.tight_layout() 
# Add a grid to the plot for better readability
plt.grid(True) 

plt.show()
```

```python
x = np.random.randn(300)
bins = 70

plt.figure(figsize=(5, 5)) 
plt.hist(x, bins,  color='r') 

plt.title("sin curve")
plt.xlabel("X axis")
plt.ylabel("Y axix")

# Add a legend to identify the plotted line
plt.legend() 
# Automatically adjust spacing to prevent overlap of labels or titles
plt.tight_layout() 
# Add a grid to the plot for better readability
plt.grid(True) 

plt.show()
```

```python
x = ["cats", "dogs", "birds"]
y = [10, 22, 5]

plt.figure(figsize=(3, 4)) 
plt.pie(y, labels = x, autopct='%1.1f%%',startangle=180, colors =['r', 'g', 'b'])   

plt.title("sin curve")
plt.xlabel("X axis")
plt.ylabel("Y axix")

# Add a legend to identify the plotted line
plt.legend() 
# Automatically adjust spacing to prevent overlap of labels or titles
plt.tight_layout() 
# Add a grid to the plot for better readability
plt.grid(True) 

plt.show()
```

```python
x = np.linspace(0, 20, 100)
y = np.sin(x)
new = np.cos(x)

plt.figure(figsize=(10, 5)) 
plt.plot(x, y, label='sin(x)', color='r') 
plt.plot(x, new, label='cos(x)', color='b') 

plt.title("sin curve")
plt.xlabel("X axis")
plt.ylabel("Y axix")

# Add a legend to identify the plotted line
plt.legend() 
# Automatically adjust spacing to prevent overlap of labels or titles
plt.tight_layout() 
# Add a grid to the plot for better readability
plt.grid(True) 

plt.show()
```
